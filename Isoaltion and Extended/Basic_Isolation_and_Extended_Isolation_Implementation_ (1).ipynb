Isoaltionforest with  parameters

```python
df=pd.read_csv('/content/winequality-white.csv')
```

```python
#giving a variable to Isoaltionforest with parameters
model = IsolationForest(n_estimators=100, max_samples=256, contamination=0.2,random_state=42)
model.fit(df)  #fiting the dataframe
pred_test = model.predict(df) #predicting the outliers(-1 are outliers and 1 are normal data points)
```

```python
decision_score=model.decision_function(df) #It returns a score with high negative score are more anomalous
```

```python
df['pred_test']=pred_test
df['decision_score']=decision_score
df[df['pred_test']==-1]  #checking only outliers
```

Extended Isolation Forest

```python
df1=pd.read_csv('/content/winequality-white.csv') #reading the dataset
```

ExtensionLevel decides what level of extension to be used

if ExtensionLevel == 0 nothing but isolation forest

ifExtensionLevel == no of features -1 ,called extended isolation forest 

```python
model1=iso.iForest(df1.values,ntrees=100, sample_size=500, ExtensionLevel=0) #if 
model2=iso.iForest(df1.values,ntrees=100, sample_size=500, ExtensionLevel=df1.shape[1]-1) #EIf
```

```python
model1.compute_paths(df1.values)  #gives anomaly score for if
```

```python
model2.compute_paths(df1.values)  #gives anomaly score for Eif
```

```python
pred_class1=np.where(model1.compute_paths(df1.values)>0.55,1,0)  #converting the values to 1 which are more than .55 and 0 which are less than
pred_class2=np.where(model2.compute_paths(df1.values)>0.55,1,0)  #converting the values to 1 which are more than .55 and 0 which are less than
```

```python
pred_class1 # Outliers for If
```

```python
pred_class2  # Outliers for EIf
```

```python
df1['pred_class1']=pred_class1  # Outliers for If
df1['pred_class2']=pred_class2  # Outliers for EIf
```

```python
df1[df1['pred_class1']==1]  # checking only outliers if 
```

```python
df1[df1['pred_class2']==1]   # checking only outliers for Eif
```
